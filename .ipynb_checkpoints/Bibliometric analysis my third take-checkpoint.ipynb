```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from sklearn.manifold import MDS
import re
import collections
```

```python
def parse_ris_file(filename):
    records = []
    current_record = {}

    with open(filename, 'r') as file:
        for line in file:
            line = line.strip()
            if line.startswith('AU  -'):  # Author
                if 'First Author' not in current_record:
                    current_record['First Author'] = line[6:].strip()
            elif line.startswith('PY  -'):  # Publication Year
                current_record['Year'] = line[6:].strip().split('/')[0]  # Assuming format 'YYYY/MM/DD/'
            elif line.startswith('AB  -'):  # Abstract
                current_record['Abstract'] = line[6:].strip()
            elif line.startswith('DO  -'):  # DOI
                current_record['DOI'] = line[6:].strip()
            elif line == 'ER  -':  # End of record
                records.append(current_record)
                current_record = {}

    return pd.DataFrame(records)

# Replace 'path_to_your_ris_file.ris' with the path to your RIS file
initial_df = parse_ris_file('velophary_endnote_library.txt')
initial_df
```

```python
responses = []

for doi in initial_df["DOI"]:
    if pd.notna(doi):
        api_call = f"https://opencitations.net/api/v1/metadata/{doi}"
        http_headers = {"authorization": "739eea95-2ca0-47b1-a11a-84028da7c324"}
        response = requests.get(api_call, headers=http_headers)
        if response.status_code == 200:
            responses.append(response.json())
        else:
            responses.append(None)
    else:
        responses.append(None)
```

```python
responses
```

```python
def element_to_df(element):
    if element is None:
        return None
    else:
        return pd.DataFrame(element)

# Apply the function to each element and concatenate the results
primary_papers = pd.concat([element_to_df(element) for element in responses], ignore_index=True)
```

```python
primary_papers = pd.read_excel("primary_papers_1205.xlsx")
```

```python
# Creating the new DataFrame with 'Citing Paper' and 'Cited Paper'
pattern = r'10\.\d+/\S+?(?=;|$)'
citations = []
for index, row in primary_papers.iterrows():
    # Check if 'References' is not NaN
    if pd.notna(row['reference']):
        # Use regular expression to extract DOIs from the 'References' string
        reference_dois = re.findall(pattern, row['reference'])
        
        # Iterate through each extracted DOI
        for cited_doi in reference_dois:
            # Append the citation to the citations list
            citations.append({'citing': row['doi'], 'cited': cited_doi})

citation_df = pd.DataFrame(citations)


# Print the citation DataFrame
citation_df
```

```python
doi_pattern = r'10.\d{4,9}/[-._;()/:A-Za-z0-9]+'

# List to store the responses
responses_secondary = []

# Iterate over each row in the DataFrame
for index, row in citation_df.iterrows():
    # Extract DOIs from the 'Cited Reference DOI' column
    cited_reference_dois = re.findall(doi_pattern, row['cited'])

    # Iterate through each extracted DOI
    for doi in cited_reference_dois:
        # API call to retrieve metadata
        api_call = f"https://opencitations.net/api/v1/metadata/{doi}"
        http_headers = {"authorization": "739eea95-2ca0-47b1-a11a-84028da7c324"}  # Replace with your actual API key
        response = requests.get(api_call, headers=http_headers)

        # Append the response JSON if the status code is 200, else append None
        if response.status_code == 200:
            responses_secondary.append(response.json())
        else:
            responses_secondary.append(None)

# responses_secondary now contains the metadata for each cited reference DOI
```

```python
secondary_papers = pd.concat([element_to_df(element) for element in responses_secondary], ignore_index=True)
secondary_papers['year'] = secondary_papers['year'].str.split('-').str[0]
secondary_papers
```

```python
# Assuming citation_df is your DataFrame and has already been defined
tuples = [tuple(x) for x in citation_df.values]

# Create a directed graph
G = nx.DiGraph()
G.add_edges_from(tuples)

# Separate the 'Citing Papers' and 'Cited Papers'
citing_papers = set(citation_df['citing'])
cited_papers = set(citation_df['cited']) - citing_papers

# Create a color map: 'red' for citing papers, 'blue' for cited papers
color_map = []
for node in G:
    if node in citing_papers:
        color_map.append('red')  # Color for primary papers
    else:
        color_map.append('blue')  # Color for secondary papers


# Draw the graph
#fig, ax = plt.subplots(figsize=6,6)
plt.figure(figsize=(6, 6))
nx.draw(G, node_size=70, node_color=color_map, alpha=0.2, font_size=8, with_labels = False)
#limits=plt.axis('on')
#ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
plt.title("Co-Citation Network")
plt.show()
```

```python
### degree_distribution viz or intranetwork citation count###

# Getting the degree (citation count) of each node
degree_sequence = [d for n, d in G.degree()]

# Node IDs (assuming they are simply numbered from 0 to n-1)
node_ids = list(range(len(degree_sequence)))

# Sorting the nodes by their degree (citation count)
sorted_nodes = sorted(G.degree, key=lambda x: x[1], reverse=True)

# Unzipping the sorted node IDs and their degrees
sorted_node_ids, sorted_degrees = zip(*sorted_nodes)

# Plotting the sorted data
plt.figure(figsize=(10, 10))
plt.scatter(range(len(sorted_node_ids)), sorted_degrees, color='r', s=5)  # Reduced size of dots

for i in range(5):
    plt.text(i, sorted_degrees[i], f'DOI {sorted_node_ids[i]}', fontsize=9, ha='left')

plt.xlabel('Node ID')
plt.ylabel('Intra-network Citation Count')
plt.title('Degree Distribution of Network Nodes')
plt.xticks(range(len(sorted_node_ids)), sorted_node_ids)
plt.show()


```

```python
meta_data = pd.concat([primary_papers, secondary_papers])
```

```python
# Initialize an empty list
doi_list = []

# Iterate over each index value (DOI) in double_sorted_df
for doi in double_sorted_df.index:
    doi_list.append(doi)

len(doi_list)
```

```python
# Initialize an empty list to store the data
years_doi_data = []

# Loop through each DOI in doi_list
for doi in doi_list:
    # Find the corresponding row in meta_data
    row = meta_data[meta_data['doi'] == doi]

    # If the doi is found in meta_data, append the 'year' and 'doi' to the list
    if not row.empty:
        year = row['year'].iloc[0]
        years_doi_data.append({'year': year, 'doi': doi})

# Create a DataFrame from the list
years_doi = pd.DataFrame(years_doi_data)

years_doi.sort_values("doi")
years_doi
```

```python
# Initialize an empty list to store the data and a list for missing DOIs
years_doi_data = []
missing_dois = []

# Loop through each DOI in doi_list
for doi in doi_list:
    # Find the corresponding row in meta_data
    row = meta_data[meta_data['doi'] == doi]

    # Check if the doi is found in meta_data
    if not row.empty:
        year = row['year'].iloc[0]
        years_doi_data.append({'year': year, 'doi': doi})
    else:
        missing_dois.append(doi)

# Create a DataFrame from the list
years_doi = pd.DataFrame(years_doi_data)

# Sort by 'doi'
years_doi = years_doi.sort_values("doi")

# Check the lengths and missing DOIs
len(years_doi), len(missing_dois)
```

```python
years_doi_list = years_doi['doi'].tolist()
filtered_bc_matrix = bc_matrix.loc[bc_matrix.index.isin(years_doi_list), bc_matrix.columns.isin(years_doi_list)]
filtered_bc_matrix
```

```python
years = [str(year) for year in years_doi['year'].tolist()]
len(years)
```

```python
# Apply MDS to reduce the bibliographic coupling matrix to one dimension
mds = MDS(n_components=1, dissimilarity="precomputed", random_state=42)
semantic_positions = mds.fit_transform(1 / (filtered_bc_matrix + 1)).flatten()  # Inverse to convert to 'distance'
type(semantic_positions)
```

```python

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot of year of publication against semantic position
plt.scatter(years, semantic_positions, color='blue', alpha=0.6)

# Labeling the axes
plt.xlabel('Year of Publication')
plt.ylabel('Semantic Position')
plt.title('Semantic Similarity vs. Year of Publication')
plt.grid(True)
plt.show()
```

```python
# Extract DOI values from the index of filtered_bc_matrix
doi_values = filtered_bc_matrix.index.tolist()

# Ensure the lengths match
assert len(doi_values) == len(semantic_positions), "Lengths of DOI values and semantic positions do not match."

# Retrieve the list of years corresponding to these DOI values
# Assuming years_doi is a DataFrame with 'doi' and 'year' columns
years = [years_doi[years_doi['doi'] == doi]['year'].iloc[0] for doi in doi_values]

# Create a new DataFrame
combined_df = pd.DataFrame({
    'Semantic_Position': semantic_positions,
    'DOI': doi_values,
    'Year': years
})

print(combined_df)
```

```python
G = nx.Graph()

# Add nodes with 'year' attribute
for index, row in dropped_.iterrows():
    G.add_node(row['DOI'], year=row['Year'])

# Add edges with weights from filtered_bc_matrix
for doi1 in dropped_filt_bc_matr.columns:
    for doi2 in dropped_filt_bc_matr.index:
        if doi1 != doi2:
            weight = dropped_filt_bc_matr.at[doi2, doi1]
            G.add_edge(doi1, doi2, weight=weight)

# Position nodes for visualization, sorting them by year
pos = {}
years = sorted(dropped_['Year'].unique())
layer_spacing = 1.0  # Adjust as needed
for node, data in G.nodes(data=True):
    year = data['year']
    layer = years.index(year) * layer_spacing
    pos[node] = (layer, -layer)  # Adjust as needed

# Draw the network
edge_weights = nx.get_edge_attributes(G, 'weight')
nx.draw(G, pos, with_labels=True, 
        edge_cmap=plt.cm.Blues, 
        edge_color=[edge_weights[edge] for edge in G.edges])
plt.show()
```
